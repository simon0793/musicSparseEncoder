```python
%matplotlib notebook
from pylab import*
from scipy.io import wavfile
import IPython
```

```python
sampFreq, snd = wavfile.read('./trainWav/mary had a little lamb.wav')
```

```python
print(snd.dtype)
```

```python
snd = snd / (2.**15)
s1 = snd[:,0] 
```

```python
timeArray = arange(0, snd.shape[0], 1)
timeArray = timeArray / sampFreq
timeArray = timeArray * 1000  #scale to milliseconds
```

```python
plot(timeArray[:10], s1[:10], color='k')
ylabel('Amplitude')
xlabel('Time (ms)')
```

```python
n = len(s1) 
p = fft(s1) # take the fourier transform 
```

```python
nUniquePts = int(ceil((n+1)/2.0))
p = p[0:nUniquePts]
p = abs(p)
```

```python
p = p / float(n) # scale by the number of points so that
                 # the magnitude does not depend on the length 
                 # of the signal or on its sampling frequency  
p = p**2  # square it to get the power 

# multiply by two (see technical document for details)
# odd nfft excludes Nyquist point
if n % 2 > 0: # we've got odd number of points fft
    p[1:len(p)] = p[1:len(p)] * 2
else:
    p[1:len(p) -1] = p[1:len(p) - 1] * 2 # we've got even number of points fft

freqArray = arange(0, nUniquePts, 1.0) * (sampFreq / n);
plot(freqArray/1000, 10*log10(p), color='k')
xlabel('Frequency (kHz)')
ylabel('Power (dB)')
```

```python
from sklearn.linear_model import LinearRegression
import random
```

```python
d = int(second / 64)
second = sampFreq
length = 5*second
X = np.fliplr([s1[i:i+d] for i in range(0,(length - d + 1))])
Y = s1[d:(length + 1)]
print(X.shape)
print(Y.shape)
```

```python
regressor = LinearRegression(normalize=True)
regressor.fit(X, Y)
```

```python
start = 25*second + 1048
end = start + 2*second
length = end - start + 1
song = np.zeros((length,))
frame = s1[start:start+d]
numRandom = 0
count = second
for i in range(count):
#     X_test = np.fliplr([s1[i:i+d] for i in range(start,(end - d + 1))])
    nextNote = regressor.predict(np.fliplr(frame[np.newaxis, :]))
    if nextNote[0] <= -1:
        numRandom += 1
        nextNote[0] = random.uniform(-1, -0.5)
    if nextNote[0] >= 1:
        numRandom += 1
        nextNote[0] = random.uniform(0.5, 1)
    song[i] = nextNote[0]
    frame = np.roll(frame, -1)
#     if i % 2 == 0:
    frame[d-1] = nextNote[0]
    if abs(nextNote[0]) < 0.001:
        numRandom += 1
        temp = start + i
        frame = s1[temp:temp+d]
print("Percent random: " + str(100*numRandom/count))
```

```python
print(song[:100])
```

```python
print(song.shape)
print(np.unique(song).shape)
generatedSong = (song * (2.**15)).astype(np.int16)
wavfile.write('linearGen.wav', sampFreq, generatedSong)
```

```python
actualSong = (s1[start:end] * (2.**15)).astype(np.int16)
wavfile.write('actual.wav', sampFreq, actualSong)
```
